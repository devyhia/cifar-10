## Yellow Pages Scraping

```python
import re
import json
import http.client
from lxml import html
import pandas as pd
import urllib
import time
```

```python
def count_per_location(city):
    conn = http.client.HTTPConnection("www.yellowpages.com")
    
    headers = {
        'cache-control': "no-cache",
        'postman-token': "ccbfc73b-775a-04f8-519e-93953e05d913"
        }
    
    load = {
        'search_terms': 'translation',
        'geo_location_terms': city
    }

    
    try:
        conn.request("GET", "/search?%s" % urllib.parse.urlencode(load), headers=headers)
        res = conn.getresponse()
        data = res.read()
        body = data.decode("utf-8")

        tree = html.fromstring(body)
        elem = tree.xpath('//div[@class="pagination"]/p/text()')
        count = int(re.compile('of\s(\d+)$').search(elem[0]).group(1))
    except Exception as e:
        print(e)
        print(body)
        print(elem)
        count = -1
    
    return count
```

```python
cities = pd.read_json("us_cities.json", orient='records')
```

```python
cities['query'] = cities.City + ", " + cities.State
```

```python
cities['count'] = 0
```

```python
cities_unique = cities.drop_duplicates(subset=['query'])
```

```python
cities_unique['count'] = cities_unique['query'].apply(count_per_location)
```

```python
states = pd.read_json('states.json', orient='records')
```

```python
states = states[states['abbreviation'] != 'AS']
states = states[states['abbreviation'] != 'FM']
states = states[states['abbreviation'] != 'MH']
states = states[states['abbreviation'] != 'MP']
states = states[states['abbreviation'] != 'PW']
```

```python
def all_cities_of(state):    
    conn = http.client.HTTPConnection("api.sba.gov")

    headers = {
        'cache-control': "no-cache",
        'postman-token': "add70860-5365-7d32-89cd-ad3bdcf968fd"
        }

    conn.request("GET", "/geodata/city_data_for_state_of/{}.json".format(state), headers=headers)

    res = conn.getresponse()
    data = res.read()

    return json.loads(data.decode("utf-8"))
```

```python
cities = []
for state in states.abbreviation.tolist():
    print("Loading {} ...".format(state), end="")
    batch = all_cities_of(state.lower())
    cities += batch
    print(" {} ... {}.".format(len(batch), len(cities)))
```

```python
cities = pd.DataFrame(cities)
```

```python
cities.state_abbreviation.unique().shape
```

```python
cities
```

```python
cities['translators_count'] = 0
for idx, city in cities.iterrows():
    cities.loc[idx, 'translators_count'] = count_per_location("{}, {}".format(city['name'], city['state_abbreviation']))
    print("\r{0:.2f}% Complete. {count} Translators.".format(((idx+1)*100/cities.shape[0]), count=cities['translators_count'].sum()), end="")
```

```python
cities.loc[cities.translators_count.sort_values(ascending=False).index][:50]
```
